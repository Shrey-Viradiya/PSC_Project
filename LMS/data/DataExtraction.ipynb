```python
import pandas as pd
```

```python
BooksDF = pd.read_csv('books.csv')
```

```python
BooksDF.columns
```

```python
authors = BooksDF['Author'].unique()
```

```python

authors = list(map(lambda x: x.title(), authors))
print(len(authors))
authors = list(set(authors))
print(len(authors))
authors.sort()
for _ in authors:
    print(_)
```

```python
with open('authors.txt','w') as file:
    for _ in authors:
        file.write(_+"\n")
```

```python
publisher = BooksDF['Publisher'].unique()
print(len(publisher))
```

```python
publisher.sort()
```

```python
def rep(x):
    return x.replace('&amp;', '&')
```

```python
publisher = list(map(rep, publisher))
```

```python
publisher
```

```python
with open('publ.txt','w') as file:
    for _ in publisher:
        file.write(_+"\n")
```
