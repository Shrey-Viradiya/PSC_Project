```python
import pandas as pd
```

```python
users = '../data/users.csv'
books = '../data/books.csv'
ratings = '../data/ratings.csv'
```

```python
usersDF = pd.read_csv(users)
```

```python
booksDF = pd.read_csv(books)
```

```python
ratingsDF = pd.read_csv(ratings)
```

```python
print(ratingsDF.columns)
```

```python
selecting_most_users = ratingsDF.groupby('userID').agg('count')
selecting_most_users = selecting_most_users.sort_values(by = ['rating'],ascending= False)
selecting_most_users = selecting_most_users.head(10000)
selecting_most_users = selecting_most_users.reset_index()
print(selecting_most_users)
```

```python
print(selecting_most_users.columns)
```

```python
usersList = list(selecting_most_users['userID'].unique())
```

```python
print(usersList)
print(len(usersList))
```

```python
print(ratingsDF.shape)
ratingsDF = ratingsDF[ratingsDF['userID'].isin(usersList)]
print(ratingsDF.shape)
```

```python
print(usersDF.shape)
usersDF = usersDF[usersDF['User-ID'].isin(usersList)]
print(usersDF.shape)
```

```python
usersDF.to_csv('users.csv')
ratingsDF.to_csv('ratings.csv')
```
